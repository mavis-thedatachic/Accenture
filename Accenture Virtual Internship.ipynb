## Merging Cleaned Data into one DataFrame

```python
merged_data = pd.merge(reaction_type,reactions,on='Type').merge(content,on='Content ID')
```

```python
merged_data
```

```python
#dup = merged_data.duplicated()
#merged_data[dup].sum()
```

```python
merged_data['Total_score']=merged_data.groupby('Category')['Score'].transform('sum')
```

```python
merged_data
```

```python
merged_data_sort = merged_data.sort_values('Total_score',ascending = False)
top = merged_data_sort.head(5).index.tolist()
```

```python
merged_data['Total_score'].max()
```

```python
merged_data_sort
```

```python
top_5 = merged_data_sort['Category'].value_counts().head(5)
```

```python
print('Top 5 Categories: ',top_5)
```

```python
print(top_5)
```

```python
top_5_data = pd.DataFrame(top_5)
```

```python
top_5_data = top_5_data.reset_index()
```

```python
top_5_data
```

```python
with pd.ExcelWriter('top5.xlsx') as writer:
    top_5_data.to_excel(writer,sheet_name='top5',index = False)
```
